```python
# deal with null vale in total charges
df['TotalCharges'] = df['TotalCharges'].replace(' ', '0')
```

```python
# now we change the data type of total charges to float
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
# we convert the senior citizen to categorical data
def convert(value):
    if value == 0:
        return 'No'
    else:
        return 'Yes'
df['SeniorCitizen'] = df['SeniorCitizen'].apply(convert)
```

```python

df.head()
# we do to make the data more readable for senior citizen
```

```python
# convert the 0 into the no and the `1` into the yes
# make the countplot for the senior citizen
ax=sns.countplot(x='SeniorCitizen', data=df)
ax.bar_label(ax.containers[0]) # add the label to the bar chart 
# add the title to the bar chart
plt.title('Senior Citizen Count')
plt.show()
```

```python

ax=sns.countplot(x='SeniorCitizen', data=df, hue='Churn')
ax.bar_label(ax.containers[0]) # add the label to the bar chart
```

```python
# if we want to see the data in the percentage
# now we group by data by the senior citizen and churn
# make the figuresize change and resize the figure
plt.figure(figsize=(5,6))
gb=df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'] , labels=gb.index, autopct='%1.1f%%')
plt.show()
```
