```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'D:\c desktop data\Python\archive (1)\train.csv')
```

```python
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# nw we split the traing anfd the test data
from sklearn.model_selection import train_test_split 
# We split the data into training and test data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# now we doing the feature scalling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test) 
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

model.compile(loss='mse', optimizer='adam', metrics=['mae'])
history = model.fit(X_train, y_train, epochs=20, batch_size=32)
```

```python
import numpy as np

# Example new mobile input data with all 20 features in correct order (excluding target 'price_range')
new_mobile = np.array([[
    842,    # battery_power
    0,      # blue
    2.2,    # clock_speed
    0,      # dual_sim
    1,      # fc (front camera megapixels)
    0,      # four_g
    7,      # int_memory (internal memory in GB)
    0.6,    # m_dep (mobile depth in cm)
    188,    # mobile_wt (mobile weight in grams)
    2,      # n_cores (number of cores)
    10,     # pc (primary camera megapixels, example)
    20,     # px_height (pixel resolution height)
    756,    # px_width (pixel resolution width)
    2549,   # ram (RAM size in MB)
    9,      # sc_h (screen height in cm)
    7,      # sc_w (screen width in cm)
    19,     # talk_time (hours)
    0,      # three_g (3G support: 0 or 1)
    0,      # touch_screen (touch screen: 0 or 1)
    1       # wifi (WiFi support: 0 or 1)
]])

# Scale the input using the same scaler used in training
new_mobile_scaled = sc.transform(new_mobile)

# Predict the price range (output is continuous)
predicted_price_range = model.predict(new_mobile_scaled)

# Round to nearest integer (0 to 3) for classification
predicted_class = np.round(predicted_price_range).astype(int)

print(f"Predicted price range class (0=low, 3=high): {predicted_class[0][0]}")
```
