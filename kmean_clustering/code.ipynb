```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
    df=pd.read_csv("Mall_Customers (2).csv")
```

```python
# spllit the datset into the x 
X=df.iloc[:,[3,4]].values
print(X)
```

### now we apply the albow method to find the number cluster 

```python
from sklearn.cluster import KMeans

# We make a list in which the WCSS values will be saved
wcss = []

# Now we compute the WCSS values to find the optimal number of clusters
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)  # Fixed typo and added random_state
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init="k-means++", random_state=42)  # Fixed typo and added random_state
y_kmean=kmeans.fit_predict(X)
print(y_kmean)
```

```python
## now we have the 5 cluster so we make the 5 catter plot
plt.scatter(X[y_kmean ==0 , 0   ], X[y_kmean ==0 , 1], s=100, c="red", label="Cluster 1")
plt.scatter(X[y_kmean ==1 , 0   ], X[y_kmean ==1 , 1], s=100, c="blue", label="Cluster 2")
plt.scatter(X[y_kmean ==2 , 0   ], X[y_kmean ==2 , 1], s=100, c="green", label="Cluster 3")
plt.scatter(X[y_kmean ==3 , 0   ], X[y_kmean ==3 , 1], s=100, c="cyan", label="Cluster 4")
plt.scatter(X[y_kmean ==4 , 0   ], X[y_kmean ==4 , 1], s=100, c="magenta", label="Cluster 5")

plt.title("Customer_segmentation")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.show()            
```
