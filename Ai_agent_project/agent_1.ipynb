```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os
import asyncio
import nest_asyncio

# Allow nested event loops (fix for Jupyter/IPython)
nest_asyncio.apply()

# Set your Gemini API key
os.environ["GOOGLE_API_KEY"] = ""  # Replace with your actual Gemini API Key

# GeminiAgent definition
class GeminiAgent:
    def __init__(self, name, instructions, model="gemini-1.5-flash", temperature=0.7):
        self.name = name
        self.instructions = instructions
        self.llm = ChatGoogleGenerativeAI(
            model=model,
            temperature=temperature
        )
    
    async def run(self, prompt):
        full_prompt = f"{self.instructions}\n\nUser: {prompt}"
        response = await self.llm.ainvoke(full_prompt)
        return response.content

# Jupyter-safe Runner
class GeminiRunner:
    @staticmethod
    def run_sync(agent, prompt):
        async def run_agent():
            return await agent.run(prompt)

        # This will now work in Jupyter thanks to nest_asyncio
        return_result = asyncio.get_event_loop().run_until_complete(run_agent())

        class Result:
            def __init__(self, output):
                self.final_output = output

        return Result(return_result)

# Example usage
agent = GeminiAgent(
    name="Assistant",
    instructions="You are a helpful assistant"
)

result = GeminiRunner.run_sync(agent, "Write a haiku about recursion in programming.")
print(result.final_output)
```
