```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns







```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
## now we are use the decision tree model

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(dataset.iloc[:, 1:2].values, dataset.iloc[:, 2].values)
```

```python
## if we want to predict the any value we can use the predict method
y_pred= regressor.predict([[6.5]])
print(y_pred)
```

```python
# now we are going to visualize the result
plt.scatter(dataset.iloc[:, 1:2].values, dataset.iloc[:, 2].values, color='red')
plt.plot(dataset.iloc[:, 1:2].values, regressor.predict(dataset.iloc[:, 1:2].values), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
