```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import matplotlib.ticker as ticker
import matplotlib.lines as mlines
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

import sklearn.metrics as metrics
```

```python
df=pd.read_csv('Fertilizers.csv')
```

```python
df
```

```python
# now we split the dataset and assin the x and the y values
X = df.iloc[:, :-1].values

y = df.iloc[:, -1].values
```

```python
## now we split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
# now trai the data by linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
# now we predict the test set results
y_pred = regressor.predict(X_test)
# now we check the accuracy of the model by using the mean absolute error, mean squared error and root mean squared error
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
# now we check the accuracy of the model by using the r2 score
from sklearn.metrics import r2_score
print('R2 Score:', r2_score(y_test, y_pred))

```

```python
# now we will plot the test set results
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred, color='blue')
plt.title('Fertilizers Test Set Results')
plt.xlabel('Fertilizers')
plt.ylabel('Yield')
plt.show()  
```

```python
# add the detail that become easy to understand in the plot
plt.scatter(X_test, y_test, color='red', label='Actual Yield')
plt.plot(X_test, y_pred, color='blue', label='Predicted Yield')
plt.title('Fertilizers Test Set Results')
plt.xlabel('Fertilizers')
plt.ylabel('Yield')
plt.legend()
plt.show()
```
