```python
### polynomial regreesion
###model:to make predictions based on polynomial regression in wich predict the salary of the new empolyee that are hire in the organization wether all the position have diff diff salary and the salary is not linear with respect to the level of the employee. so in case 
### of that we predict the value of the new empolyee salary based on the level of the employee.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
### now importing thee dataset
dataset = pd.read_csv('Position_Salaries.csv')
```

## train the linear regression model on whole dataset

```python
x =  dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
### now we will split the dataset into training and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# now apply the linear regression model
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
LinearRegression()
```

```python
# now we visualize the linear regression model
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, regressor.predict(x_train), color='blue')
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()

```

```python
#now we tain the polynomail regression model
# x convert int ot the polynomial featuresimport numpy as np
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg=PolynomialFeatures(degree=4)
```

```python
# now we apply on our dataset
x_poly=poly_reg.fit_transform(x_train)
# now we apply the linear regression model on the polynomial features
```

```python
print(x_poly)
```

```python
regressor_1=LinearRegression()
regressor_1.fit(x_poly, y_train)
```

```python
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, regressor_1.predict(x_poly), color='blue')
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# Train Polynomial Regression model
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x_train)

poly_model = LinearRegression()
poly_model.fit(x_poly, y_train)

# For smooth curve
import numpy as np

x_grid = np.arange(min(x_train), max(x_train), 0.1)
x_grid = x_grid.reshape(-1, 1)

# Plotting Polynomial Regression results
plt.scatter(x_train, y_train, color='red')
plt.plot(x_grid, poly_model.predict(poly_reg.transform(x_grid)), color='blue')
plt.title('Salaries based on levels')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```
